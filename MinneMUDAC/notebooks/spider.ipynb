### Historical Futures Prices for Soybean: 2000-2018

- empty_url: urls that don't exist
- miss_url: urls that don't have complete soybean data

```python
## crawl soybean data for 2000-2018
soybean = []
soybean_meal = []
soybean_oil = []
empty_url = []
miss_url = []
for year in np.arange(2000, 2019):
    for month in np.arange(1, 13):
        for day in np.arange(1, 32):
            url = "https://www.mrci.com/ohlc/{}/{}.php".format(year, str(year)[2:4]+str(month).zfill(2)+str(day).zfill(2))
            r = requests.get(url)
            if not r:
                empty_url.append(url)
                continue
            soup = BeautifulSoup(r.content)
            table = soup.find('table', attrs={'border':'3', 'cellspacing': '1', 'cellpadding':1})
            rows = table.findAll('tr')
            idxh = [i for i, item in enumerate(rows) if item.text.strip().startswith('Soybean')]
            idxt = [i for i, item in enumerate(rows) if item.text.strip().startswith('Total')]
            if len(idxh)<3:
                miss_url.append(url)
                continue
            rows_1 = rows[(idxh[0]+1):(idxt[0])]  # get the rows for Soybeans(CBOT)
            rows_2 = rows[(idxh[1]+1):(idxt[1])]  # get the rows for Soybean Meal(CBOT)
            rows_3 = rows[(idxh[2]+1):(idxt[2])]  # get the rows for Soybean Oil(CBOT)

            for row in rows_1:
                cols = row.findAll('td')
                cols = [ele.text.strip() for ele in cols]
                if cols:
                    soybean.append([ele for ele in cols if ele])
                    
            for row in rows_2:
                cols = row.findAll('td')
                cols = [ele.text.strip() for ele in cols]
                if cols:
                    soybean_meal.append([ele for ele in cols if ele])

            for row in rows_3:
                cols = row.findAll('td')
                cols = [ele.text.strip() for ele in cols]
                if cols:
                    soybean_oil.append([ele for ele in cols if ele])
        print("{}/{}".format(year, str(month).zfill(2)))
```

```python
## write soybean data to csv for 2000-2018
with open ('soybean.csv','w') as file:
    writer=csv.writer(file)
    for row in soybean:
        writer.writerow(row)

with open ('soybean_meal.csv','w') as file:
    writer=csv.writer(file)
    for row in soybean_meal:
        writer.writerow(row)
        
with open ('soybean_oil.csv','w') as file:
    writer=csv.writer(file)
    for row in soybean_oil:
        writer.writerow(row)
```

```python
## write empty and missing urls for 2000-2018
with open ('empty_url_hist.md', 'w') as file:
    for row in empty_url:
        file.write('{}\n'.format(row))

with open ('miss_url_hist.md', 'w') as file:
    for row in miss_url:
        file.write('{}\n'.format(row))
```

## Futures Prices for 2019

Commodities: soybean, soybean meal, soybean oil, corn, rough rice, lean hogs, conola

```python
# crawl for all considered commodities for 2019 year
com_list = ['Soybeans(CBOT)', 'Soybean Meal(CBOT)', 'Soybean Oil(CBOT)', 'Corn', 'Rough Rice', 'Lean Hogs', 'Canola']
result_list = [[None]]*7
empty_url = []
year = 2019
for month in np.arange(1, 13):
    for day in np.arange(1, 32):
        url = "https://www.mrci.com/ohlc/{}/{}.php".format(year, str(year)[2:4]+str(month).zfill(2)+str(day).zfill(2))
        r = requests.get(url)
        if not r:
            empty_url.append(url)
            continue
        soup = BeautifulSoup(r.content)
        table = soup.find('table', attrs={'border':'3', 'cellspacing': '1', 'cellpadding':1})
        rows = table.findAll('tr')
     
        for k in range(len(com_list)):            
            # get rows for each commodity
            idxh = [i for i, item in enumerate(rows) if item.text.strip().startswith(com_list[k])][0]
            idxt = [i for i, item in enumerate(rows) if (item.text.strip().startswith('Total') and i > idxh)][0]
            rows_com = rows[(idxh+1):(idxt)]  
            
            for row in rows_com:
                cols = row.findAll('td')
                cols = [ele.text.strip() for ele in cols]
                if cols:
                    result_list[k].append([ele for ele in cols if ele])       
            
    print("{}/{}".format(year, str(month).zfill(2)))
```

```python
## write commodity data to csv for 2019
name_list = ['soybean', 'soybean_meal', 'soybean_oil', 'corn', 'rice', 'hogs', 'canola']

for k in range(len(name_list)):
    with open('{}2019.csv'.format(name_list[k]), 'w') as file:
        writer = csv.writer(file)
        for row in result_list[k]:
            writer.writerow(row)

with open ('empty_url2019.md', 'w') as file:
    for row in empty_url:
        file.write('{}\n'.format(row))
```
